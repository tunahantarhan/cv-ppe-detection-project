```python
import yaml
import shutil
import os

yaml_source = None
dataset_root = None

for root, dirs, files in os.walk('/kaggle/input'):
    if 'data.yaml' in files:
        yaml_source = os.path.join(root, 'data.yaml')
        dataset_root = root # train ve valid klasörlerinin olduğu ana dizin
        break

if yaml_source is None:
    print("HATA: data.yaml bulunamadı.")
else:
    print(f"Dosya bulundu: {yaml_source}")
    
    yaml_dest = '/kaggle/working/data.yaml'
    shutil.copy(yaml_source, yaml_dest)
    
    with open(yaml_dest, 'r') as f:
        data = yaml.safe_load(f)
        
    data['train'] = os.path.join(dataset_root, 'train/images')
    val_dir = 'valid/images' if os.path.exists(os.path.join(dataset_root, 'valid/images')) else 'val/images'
    data['val'] = os.path.join(dataset_root, val_dir) 
    
    if os.path.exists(os.path.join(dataset_root, 'test/images')):
        data['test'] = os.path.join(dataset_root, 'test/images')
        
    with open(yaml_dest, 'w') as f:
        yaml.dump(data, f)
        
    print("YAML yolları sisteme uygun bir şekilde ayarlandı.")
```
